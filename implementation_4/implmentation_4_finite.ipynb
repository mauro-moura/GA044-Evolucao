```python
casos = {
    "dominancia": {
        "wAA": 1,
        "wAa": 1,
        "waa": 0.9
    },
    "codominancia": {
        "wAA": 1,
        "waa": 0.9,
        "wAa": 0.95 # (wAA + waa) / 2
    },
    "sobredominancia": {
        "wAA": 0.9,
        "wAa": 1,
        "waa": 0.9
    },
    "desvantagem_heterozigoto": {
        "wAA": 1,
        "wAa": 0.9,
        "waa": 1
    }
}
```
